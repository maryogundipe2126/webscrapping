```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=machine+learning&txtLocation='
```

```python
page = requests.get(url).text
```

```python
soup = BeautifulSoup(page, 'lxml')
```

```python
result = soup.select('li h3')
comp_name = [r.text.strip('\r\n').replace('\r\n', '').lstrip().strip('(More Jobs)') for r in result]
comp_name
```

```python
result = soup.select('li h3')
for r in range(len(result)):
    print(result[r].text.rstrip().strip('(More Jobs)'))
```

```python
company_name = []
result = soup.select('li h3')
for r in range(len(result)):
    company_name.append(result[r].text.rstrip().strip('(More Jobs)').strip('\r\n'))
company_name
```

```python
result = soup.select('li span[class="srp-skills"]')
skills = [r.text.replace('\r\n', '').strip() for r in result]
skills
```

```python
result = soup.select('ul[class="list-job-dtl clearfix"]')
job_descrip = [r.text.split('\n\n')[1].strip().replace('\r\n', '' ).replace('Job Description:', '') for r in result]
job_descrip
```

```python
result = soup.select('ul[class="top-jd-dtl clearfix"] span')
result
```

```python
location1 = [r.text.split()[-1] for r in result]
location1
```

```python
result = soup.select('ul[class="top-jd-dtl clearfix"]')
result
```

```python
location = [r.text.split()[-1] for r in result]
location
```

```python
experience = soup.select('ul[class="top-jd-dtl clearfix"]')
experience
```

```python
year_exp = [r.text.split('\n')[1].replace('card_travel', '') for r in experience]
year_exp
```

```python
page_info = list(zip(company_name,skills,job_descrip,location,year_exp))
page_info
```

```python
page_df = pd.DataFrame(page_info,  columns=['comp_name','key_skills', 'job_description', 'location', 'years of exp'])
page_df
```
